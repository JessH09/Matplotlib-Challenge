### Analysis
Based on the statistical analysis, Ramicane was most effective at keeping a smaller average tumor size. It's mean (40.216745) was the smallest out of the data set. It's SEM (0.320955) was also the smallest suggesting that the sample is a more accurate estimate of the population mean; Capomulin is the second most effective in terms of average tumor size and SEM. We do however have to take into consideration that Capomulin and Ramicane had the highest number of regimen counts compared to the other drug regimens so their sample size was bigger allowing for more data. Even with this, Ramicane has less regimen counts (by 2) than Capomulin and still outperforms. When analyzing Capomulin further, we can see that there's a positive correlation between mouse weight (g) and tumor size. 
 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
combined_data= pd.merge(mouse_metadata,study_results, how='outer')

# Display the data table for preview
combined_data.head()
```

```python
# Checking the number of mice.
mice_total= len(combined_data['Mouse ID'].value_counts())
mice_total
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
mouseid_timepoint= combined_data.loc[combined_data.duplicated(subset =['Mouse ID', 'Timepoint']),'Mouse ID'].unique()
mouseid_timepoint
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
mouseid_timepoint_df= combined_data.loc[combined_data['Mouse ID']== 'g989']
mouseid_timepoint_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_df=combined_data[combined_data['Mouse ID'].isin(mouseid_timepoint)==False]
clean_df
```

```python
# Checking the number of mice in the clean DataFrame.
new_mice_total= len(clean_df['Mouse ID'].value_counts())
new_mice_total
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
mean = clean_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
median= clean_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
variance=clean_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
standard_deviation=clean_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
SEM=clean_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()


# Assemble the resulting series into a single summary DataFrame.
summary_df= pd.DataFrame({'Mean Tumor Volume': mean, 'Median Tumor Volune':median, 'Variance in Tumor Volume': variance,
                         'Standard Deviation in Tumor Vol': standard_deviation,'SEM in Tumor Vol':SEM})
summary_df
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)
# Using the aggregation method, produce the same summary statistics in a single line
summary = clean_df.groupby(['Drug Regimen'])[['Tumor Volume (mm3)']].agg(['mean', 'median', 'var', 'std', 'sem'])

summary
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen

correlation = st.pearsonr(average_tumor_volume.index, average_tumor_volume)
correlation
```

```python
(slope, intercept,rvalue, pvalue, stderr)= linregress(average_tumor_volume.index,average_tumor_volume)
regress_values=average_tumor_volume.index* slope + intercept
line_eq= f"y = {round(slope, 2)} x + {round(intercept, 2)}"

plt.scatter(average_tumor_volume.index,average_tumor_volume,color='b')
plt.plot(average_tumor_volume.index, regress_values, color='red')
plt.annotate(line_eq,(20,36), fontsize=14)
plt.xlabel("Weight (g)")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Weight vs Tumor Volume for Capomulin")
print(f"The r-squared is: {round(rvalue**2,3)}")
plt.show()
```
