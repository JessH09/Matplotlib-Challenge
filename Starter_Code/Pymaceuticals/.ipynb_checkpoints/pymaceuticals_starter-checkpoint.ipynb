```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
combined_data= pd.merge(mouse_metadata,study_results, how='outer')

# Display the data table for preview
combined_data.head()
```

```python
# Checking the number of mice.
mice_total= len(combined_data['Mouse ID'].value_counts())
mice_total
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
mouseid_timepoint= combined_data.loc[combined_data.duplicated(subset =['Mouse ID', 'Timepoint']),'Mouse ID'].unique()
mouseid_timepoint
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
mouseid_timepoint_df= combined_data.loc[combined_data['Mouse ID']== 'g989']
mouseid_timepoint_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_df=combined_data[combined_data['Mouse ID'].isin(mouseid_timepoint)==False]
clean_df
```

```python
# Checking the number of mice in the clean DataFrame.
new_mice_total= len(clean_df['Mouse ID'].value_counts())
new_mice_total
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
mean = clean_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
median= clean_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
variance=clean_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
standard_deviation=clean_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
SEM=clean_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()


# Assemble the resulting series into a single summary DataFrame.
summary_df= pd.DataFrame({'Mean Tumor Volume': mean, 'Median Tumor Volune':median, 'Variance in Tumor Volume': variance,
                         'Standard Deviation in Tumor Vol': standard_deviation,'SEM in Tumor Vol':SEM})
summary_df
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)
# Using the aggregation method, produce the same summary statistics in a single line
summary = clean_df.groupby(['Drug Regimen'])[['Tumor Volume (mm3)']].agg(['mean', 'median', 'var', 'std', 'sem'])

summary
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of rows (Mouse ID/Timepoints) for each drug regimen using Pandas.
regimen_counts = clean_df.groupby('Drug Regimen').size()
regimen_counts
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Start by getting the last (greatest) timepoint for each mouse
# Step 1: Group the data by mouse ID and treatment regimen
grouped_data = clean_df.groupby(['Mouse ID', 'Drug Regimen'])

# Step 2: Find the maximum timepoint for each mouse within each treatment regimen
max_timepoints = grouped_data['Timepoint'].max()

# Step 3: Retrieve the corresponding tumor volume for each mouse at the maximum timepoint
final_tumor_volumes = grouped_data.apply(lambda x: x.loc[x['Timepoint'] == max_timepoints[x.name]])

# Display the final tumor volumes
print(final_tumor_volumes[['Mouse ID', 'Drug Regimen', 'Timepoint', 'Tumor Volume (mm3)']])

```

```python
# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint

last_timepoint_merge= pd.merge(clean_df, final_tumor_volumes, how ='inner', on= 'Tumor Volume (mm3)')
last_timepoint_merge
```
